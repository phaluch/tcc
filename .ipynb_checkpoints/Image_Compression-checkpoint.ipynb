```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
```

```python
B=8 #blocksize
fn3= 'testdct.png' #path da imagem
img1 = cv.imread(fn3, cv.IMREAD_GRAYSCALE)
h,w=np.array(img1.shape[:2])/B * B #cortando imagem para um multiplo de B (tamanho do bloco)
h,w=int(h), int(w)
print(h)
print(w)
img1=img1[:h,:w]
```

```python
plt.imshow(img1[100:300, 400:600])
```

```python
blocksV=int(h/B)
blocksH=int(w/B)
vis0 = np.zeros((blocksV*B,blocksH*B))
Trans = np.zeros((blocksV*B,blocksH*B))
#print(blocksV*B,blocksH*B)
#print(img1.shape)
vis0[:blocksV*B,:blocksH*B] = img1[:blocksV*B,:blocksH*B]
for row in range(blocksV):
        for col in range(blocksH):
                #print(type(row), type(col), type(B))
                currentblock = cv.dct(vis0[row*B:(row+1)*B,col*B:(col+1)*B])
                Trans[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
plt.imshow(Trans[100:300, 400:600])
```

```python
back0 = np.zeros((blocksV*B,blocksH*B))
for row in range(blocksV):
        for col in range(blocksH):
                currentblock = cv.dct(Trans[row*B:(row+1)*B,col*B:(col+1)*B], 'Inverse')
                back0[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
plt.imshow(back0)
```

```python
diff=back0-vis0
print(diff.max())
print(diff.min())
MAD=np.sum(np.abs(diff))/float(blocksV*B*blocksH*B)
print("Mean Absolute Difference: ",MAD)
plt.figure()
plt.imshow(back0,cmap="gray")
plt.title("Backtransformed Image")
plt.show()
```
