### Compressao de Imagem padrao JPEG

```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
```

```python
B=8 #blocksize
pathToImage = 'testdct.png' #path da imagem
sourceImage = cv.imread(pathToImage, cv.IMREAD_GRAYSCALE)
h,w = np.array(sourceImage.shape[:2])/B * B #cortando imagem para um multiplo de B (tamanho do bloco)
h,w = int(h), int(w)
print(h)
print(w)
sourceImage = sourceImage[:h,:w]
```

```python
cv.imwrite('initial.jpeg', sourceImage) #gravando imagem original
```

```python
#mostrando um pedaço da imagem para testar a leitura
plt.imshow(sourceImage[100:400, 400:700],cmap="gray")
```

```python
blocksV = int(h/B) #definindo numero de blocos na vertical
blocksH = int(w/B) #definindo numero de blocos na horizontal

sourceVis = np.zeros((blocksV*B,blocksH*B)) #matriz "holder" de zeros para a visualizacao da imagem original
transformed = np.zeros((blocksV*B,blocksH*B)) #matriz "holder" de zeros para a imagem transformada

sourceVis[:blocksV*B,:blocksH*B] = sourceImage[:blocksV*B,:blocksH*B] #copia da imagem original com o tamanho correto

#aplicacao da DCT em cada bloco de BxB da imagem
for row in range(blocksV):
        for col in range(blocksH):
                currentblock = cv.dct(sourceVis[row*B:(row+1)*B,col*B:(col+1)*B])
                transformed[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
#mostrando um pedaço da imagem para testar a aplicacao da DCT
plt.imshow(transformed[100:400, 400:700], cmap="gray")
```

```python
backTransformed = np.zeros((blocksV*B,blocksH*B)) #matriz "holder" de zeros para a guardar a inversa da DCT

#aplicando a inversa da DCT com o objetivo de ter a imagem original
for row in range(blocksV):
        for col in range(blocksH):
                currentblock = cv.idct(transformed[row*B:(row+1)*B,col*B:(col+1)*B])
                backTransformed[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
#mostrando um pedaço da imagem para testar a aplicacao da inversa da DCT esperando a imagem original
plt.imshow(backTransformed[100:400, 400:700],cmap="gray")
```

```python
#mostrando a Mean Absolute Difference para checar se há divergencias entre a inversa e a original

diff = backTransformed - sourceVis
print(round(diff.max()))
print(round(diff.min()))

MAD = round(np.sum(np.abs(diff)) / float(blocksV*B*blocksH*B))
print("Mean Absolute Difference: ",MAD)

plt.figure()
plt.imshow(backTransformed,cmap="gray")
plt.title("Backtransformed Image")
plt.show()
```

```python
#mostrando a diferença visual entre a inversa e a original que nesse caso deve ser nenhuma
plt.imshow(diff,cmap="gray", vmin=0, vmax=255)
```

```python
#definindo matriz de quantizacao seguindo o padrao JPEG com fator de compressao de 50%
qMatrix = [
    [16, 11, 10, 16, 24, 40, 51, 61],
    [12, 12, 14, 19, 26, 58, 60, 55],
    [14, 13, 16, 24, 40, 57, 69, 56],
    [14, 17, 22, 29, 51, 87, 80, 62],
    [18, 22, 37, 56, 68, 109, 103, 77],
    [24, 35, 55, 64, 81, 104, 113, 92],
    [49, 61, 78, 87, 103, 121, 120, 101],
    [72, 92, 95, 98, 112, 100, 103, 99]
]
```

```python
#funcao de quantizacao
def quantize(block):
    for i in range(B):
        for j in range(B):
            block[i][j] = int(round(block[i][j] / qMatrix[i][j]))
    return block
```

```python
#funcao de dequantizacao
def dequantize(block):
    for i in range(B):
        for j in range(B):
            block[i][j] = int(round(block[i][j] * qMatrix[i][j]))
    return block
```

```python
quantized = np.zeros((blocksV*B,blocksH*B)) #matriz "holder" de zeros para a guardar a matriz transformada quantizada

#aplicando a quantizacao para cada bloco da imagem
#ONDE OCORRE O DATA LOSS
for row in range(blocksV):
        for col in range(blocksH):
                currentblock = quantize(transformed[row*B:(row+1)*B,col*B:(col+1)*B])
                quantized[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
#mostrando a imagem quantizada e transformada
plt.imshow(quantized,cmap="gray")
```

```python
#matriz "holder" de zeros para a guardar a matriz transformada e dequantizada
dequantized = np.zeros((blocksV*B,blocksH*B))

#aplicando a dequantizacao para cada bloco da imagem
for row in range(blocksV):
        for col in range(blocksH):
                currentblock = dequantize(transformed[row*B:(row+1)*B,col*B:(col+1)*B])
                dequantized[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
#mostrando a imagem dequantizada e transformada
plt.imshow(dequantized,cmap="gray")
```

```python
#matriz "holder" de zeros para a guardar a matriz destransformada e dequantizada
finalResult = np.zeros((blocksV*B,blocksH*B))

#aplicando a inversa da dct em cada bloco
for row in range(blocksV):
        for col in range(blocksH):
                currentblock = cv.idct(dequantized[row*B:(row+1)*B,col*B:(col+1)*B])
                finalResult[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
```

```python
#Calculando Mean Absolute Difference da imagem original com a final
diff1 = finalResult - sourceVis
print(round(diff1.max()))
print(round(diff1.min()))

MAD = round(np.sum(np.abs(diff1)) / float(blocksV*B*blocksH*B))
print("Mean Absolute Difference: ",MAD)
```

```python
#mostrando o resultado final da imagem
plt.imshow(finalResult,cmap="gray")
```

```python
#gravando resultado final
cv.imwrite('result.jpeg', finalResult)
```

```python
def zigZag(block):
  lines = [[] for i in range(8+8-1)] 
  for y in range(8): 
    for x in range(8): 
      i = y + x 
      if(i % 2 == 0): 
          lines[i].insert(0,block[y][x]) 
      else:  
          lines[i].append(block[y][x]) 
  return np.array([coefficient for line in lines for coefficient in line])
```

```python
import huffman
import math

from collections import Counter
def huffmanCounter(zigZagArr):
  rbCount=[]
  run=0
  for AC in zigZagArr[1:]:
    if(AC!=0):
      AC=max(AC,1-2**(2**bitBits-1)) if AC<0 else min(AC,2**(2**bitBits-1)-1)
      if(run>2**runBits-1):
        runGap=2**runBits
        k=run//runGap
        for i in range(k):
          rbCount.append('1'*runBits+'0'*bitBits)
        run-=k*runGap
      run=min(run,2**runBits-1) 
      bitSize=min(int(math.ceil(math.log(abs(AC)+0.000000001)/math.log(2))),2**bitBits-1)
      rbCount.append(format(run<<bitBits|bitSize,'0'+str(rbBits)+'b'))
      run=0
    else:
      run+=1
  rbCount.append("0"*(rbBits))
  return Counter(rbCount)
```
